```python
import numpy as np
from scipy.fftpack import fft, ifft
import pylab as plt

dt = 0.001
N = 2096
df = 1.0/(N*dt)

time = np.linspace(0.0, N*dt, N)
frequency = np.linspace(0.0, 1.0/dt, N)

wave = np.sin(time/(0.1*N*dt)*2.0*np.pi)
fourier = fft(x = wave, n = N)

plt.plot(time, wave)
```

```python
plt.xlim(xmax = 100)
plt.plot(frequency, np.abs(fourier))
```
